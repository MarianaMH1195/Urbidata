##  Flujos 

Usa el archivo analysis.py

```python
flujos = analysis.get_flujos() 

if not flujos:
    print("Atencion: No encontre el archivo CSV, asi que usare datos de prueba:")
    flujos = [
        {'origen': 'Sevilla', 'destino': 'Dos Hermanas', 'viajes': 15000, 'id_origen': '41'},
        {'origen': 'Sevilla', 'destino': 'Utrera', 'viajes': 8000, 'id_origen': '41'},
        {'origen': 'Málaga', 'destino': 'Marbella', 'viajes': 12000, 'id_origen': '29'},
        {'origen': 'Málaga', 'destino': 'Fuengirola', 'viajes': 9000, 'id_origen': '29'}
    ]

if flujos:
    print(f"Hay {len(flujos)} flujos listos para pintar.")
```

### Dibujar el mapa
líneas de los viajes.

```python
# 1. Creamos el mapa
m = folium.Map(location=[37.3891, -5.9845], zoom_start=8, tiles='CartoDB positron')

# 2. Ponemos las líneas
for f in flujos:
    ori, des, n = f['origen'].upper(), f['destino'].upper(), f['viajes']
    c1, c2 = COORDENADAS_CIUDADES.get(ori), COORDENADAS_CIUDADES.get(des)
    
    if c1 and c2:
        folium.PolyLine(
            locations=[c1, c2],
            weight=max(1, n/500),
            color='blue',
            opacity=0.5,
            popup=f"{ori} -> {des}: {n} viajes"
        ).add_to(m)
        
        folium.CircleMarker(location=c1, radius=3, color='black', fill=True, fill_color='white').add_to(m)

print("Mapa generado con exito")
m
```

```python
m.save("mapa_lineas.html")
import webbrowser, os
webbrowser.open("file://" + os.path.abspath("mapa_lineas.html"))
#se abre automáticamente el mapa generado en html para poder verlo
```
