### 1. Preparación del Entorno

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
from pathlib import Path

# Configuraciones de estilo
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("✅ Librerías cargadas")
```

### 2. Carga de Datos
Usamos el archivo procesado por `cleaning.py`.

```python
# Definimos rutas (ajustadas para el notebook)
PROCESSED_DIR = Path('data/processed')
ruta_parquet = PROCESSED_DIR / 'datos_limpios_provincias.parquet'

if ruta_parquet.exists():
    df = pd.read_parquet(ruta_parquet)
    print(f"✅ Datos cargados: {df.shape[0]} filas")
else:
    print("❌ No se encontró el archivo Parquet. Ejecuta primero cleaning.py")
```

### 3. Muestreo (Sampling)
Para que el Notebook sea fluido y los gráficos no tarden minutos, usamos una muestra de 100,000 filas.

```python
if len(df) > 100000:
    df_sample = df.sample(n=100000, random_state=42).copy()
    print("✅ Muestra de 100,000 filas generada")
else:
    df_sample = df.copy()

# Aseguramos que 'viajes' sea numérico
df_sample['viajes'] = pd.to_numeric(df_sample['viajes'], errors='coerce')
```
